```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('2019.csv')
```

```python
df.head()
```

```python
top10 = df.nlargest(10, 'Score')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x='Score', y='Country or region', palette='viridis')
plt.title('Top 10 Happiest Countries (2019)')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP per capita', y='Score', color='steelblue')

for i, row in df.head(5).iterrows():
    plt.text(row['GDP per capita'], row['Score'], row['Country or region'], fontsize=8)

for i, row in df.tail(5).iterrows():
    plt.text(row['GDP per capita'], row['Score'], row['Country or region'], fontsize=8)

plt.title('GDP per Capita vs Happiness Score (2019)')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# INSIGHT: There is a clear positive correlation between GDP and Happiness
# As GDP per capita increases, happiness score tends to increase too
# This means wealthier countries are generally happier

# INSIGHT: Low GDP countries (0.00 - 0.25) cluster around score 3.0 - 4.7
# These are likely developing/poor nations in Africa and South Asia

# INSIGHT: High GDP countries (1.25 - 1.75) cluster around score 6.0 - 7.8
# These are likely Western European and North American nations

# INSIGHT: Not a perfect relationship some high GDP countries score lower
# Money helps but doesn't guarantee happiness
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Freedom to make life choices', y='Score', color='green')
plt.title('Freedom vs Happiness Score (2019)')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# INSIGHT 1: Freedom has a positive correlation with happiness
# Countries with more freedom tend to score higher in happiness

# INSIGHT 2: Low freedom countries (0.00 - 0.10) score between 2.8 - 3.6
# These are likely authoritarian or heavily restricted nations

# INSIGHT 3: High freedom countries (0.50 - 0.65) mostly score above 5.5
# More freedom = more happiness in general

# INSIGHT 4: Freedom correlation is weaker than GDP correlation
# The scatter is more spread out compared to the GDP plot
# This means freedom matters but GDP matters more for happiness

# INSIGHT 5: Some high freedom countries still score low (~4.7)
# Freedom alone doesn't guarantee happiness without economic stability
```
